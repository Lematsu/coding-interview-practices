# Test Jusitification

Given an array of words and a width maxWidth, format the text such that each line has exactly maxWidth characters and is fully (left and right) justified.

You should pack your words in a greedy approach; that is, pack as many words as you can in each line. Pad extra spaces ' ' when necessary so that each line has exactly maxWidth characters.

Extra spaces between words should be distributed as evenly as possible. If the number of spaces on a line do not divide evenly between words, the empty slots on the left will be assigned more spaces than the slots on the right.

For the last line of text, it should be left justified and no extra space is inserted between words.

Note:

A word is defined as a character sequence consisting of non-space characters only.
Each word's length is guaranteed to be greater than 0 and not exceed maxWidth.
The input array words contains at least one word.

```
Example 1:
Input:
words = ["This", "is", "an", "example", "of", "text", "justification."]
maxWidth = 16
Output:
[
   "This    is    an",
   "example  of text",
   "justification.  "
]

Example 2:
Input:
words = ["What","must","be","acknowledgment","shall","be"]
maxWidth = 16
Output:
[
  "What   must   be",
  "acknowledgment  ",
  "shall be        "
]
```
Explanation: Note that the last line is `"shall be    "` instead of `"shall     be"`,
             because the last line must be left-justified instead of fully-justified.
             Note that the second line is also left-justified becase it contains only one word.
```
Example 3:
Input:
words = ["Science","is","what","we","understand","well","enough","to","explain",
         "to","a","computer.","Art","is","everything","else","we","do"]
maxWidth = 20
Output:
[
  "Science  is  what we",
  "understand      well",
  "enough to explain to",
  "a  computer.  Art is",
  "everything  else  we",
  "do                  "
]
```

## Communication

We could approach this problem in a greedy matter where we would combine all possible words concatenated by one whitespace within the maxWidth range and once we cannot add any more words we could incrementally add white space from the left to right in a balanced matter until we reach the max width size. We do have an exception for filling whitespace for the last line since the last line must be left justified and not fully justified like the others. The time complexity of this approach would be O(n* w) where w is the necessary whitespace needed to be added. The space complexity is O(n / m) or O(n) where n is the number of words and m is the max width size.

````python
class Solution(object):
    def fullJustify(self, words, maxWidth):
        """
        :type words: List[str]
        :type maxWidth: int
        :rtype: List[str]
        """
        res = []
        line_size = 0
        line = []
        for i, w in enumerate(words):
            line.append(w)
            line_size += len(w)
            if i == len(words) - 1 or line_size + len(words[i+1]) + len(line) > maxWidth:
                # left justified
                if i == len(words) - 1:
                    w_length = len(line)
                    whitespace = [1] * w_length
                    while line_size + sum(whitespace) > maxWidth:
                        whitespace.pop(-1)
                    while line_size + sum(whitespace) < maxWidth:
                        whitespace[w_length-1] += 1
                # full justified
                else:
                    w_index = 0
                    w_length = len(line) - 1 if len(line) > 1 else 1
                    whitespace = [1] * w_length
                    while line_size + sum(whitespace) > maxWidth:
                        whitespace.pop(-1)
                    while line_size + sum(whitespace) < maxWidth:
                        whitespace[w_index] += 1
                        w_index = (w_index + 1) % w_length
                # process
                s = ''
                for index, l in enumerate(line):
                    s += l
                    if whitespace != []:
                        s += ' ' * whitespace.pop(0)
                res.append(s)
                line = []
                line_size = 0
        return res
    def _check_results(self, l1, l2):
        if len(l1) != len(l2):
            return False
        _len = len(l1)
        for i in range(_len):
            if l1[i] != l2[i]:
                return False
                
        return True
    def unit_tests(self):
        test_cases = [
            [["What","must","be","acknowledgment","shall","be"], 16, ["What   must   be","acknowledgment  ","shall be        "]],
            [["This", "is", "an", "example", "of", "text", "justification."], 16, ["This    is    an","example  of text","justification.  "]],
            [["ask","not","what","your","country","can","do","for","you","ask","what","you","can","do","for","your","country"], 16, ["ask   not   what","your country can","do  for  you ask","what  you can do","for your country"]],
            [["Listen","to","many,","speak","to","a","few."], 6, ["Listen","to    ","many, ","speak ","to   a","few.  "]]
        ]
        for index, tc in enumerate(test_cases):
            output = self.fullJustify(tc[0], tc[1])
            assert self._check_results(output, tc[2]), 'test#{0} failed'.format(index)
            print('test#{0} passed'.format(index))
Solution().unit_tests()
                
````
